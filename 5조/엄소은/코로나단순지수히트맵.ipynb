## 코로나 단순지수

```python
jisu = pd.read_csv("/content/drive/MyDrive/8th_EDA/엄소은/전국_코로수단순피해지수.csv")
```

```python
jisu = jisu.drop(['Unnamed: 9','Unnamed: 10'],axis=1)
```

```python
jisu
```

```python
jisu = jisu.set_index('지역, 업종별 코로나피해지수')
```

```python
jisu_nosport = jisu.drop(['스포츠'],axis=1)
```

```python
jisu_nosport.columns = ['tour/leisure','real estate','service','retail','housing','food','education']
jisu_nosport['지역,업종별 코로나피해지수'] = ['Seoul','Kyungki','Daejun','Sejong','Chungbuk','Chungnam','Kyungnam','Kyungbuk','Busan','Ulsan','Incheon','Kangwon','Jeju']
```

```python
jisu_nosport = jisu_nosport.set_index('지역,업종별 코로나피해지수')
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
plt.title("Covid-19 Damage Rate")
sns.heatmap(jisu_nosport,linewidths=0.5,cmap="RdBu_r")
plt.show()
```
