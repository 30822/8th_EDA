```python
def sizing(size):
  if size[-1] == 'M':
    return float(size[:-1])
  elif size[-1] == 'k':
    return float(size[:-1])/1024
  elif size == '1,000+':
    return 1.0
  elif size == '5,000+':
    return 5.0
  elif size == '10,000+':
    return 10.0
  else:
    return np.nan
```

```python
import sys
sys.setrecursionlimit(10**7)
```

```python
data1['Size'] = data1['Size'].apply(sizing)
data1.head()
```

```python
def ratings(rating):
  if rating > 5:
    return np.nan
  elif rating >= 0:
    return rating
  else:
    return np.nan
```

```python
data1['Rating'] = data1['Rating'].apply(ratings)
```

```python
# 전처리 3: 가격 숫자 처리
```

```python
def pricing(price):
  if price[0] == '$':
    return round(float(price[1:]))
  else:
    return 0.0
```

```python
data1['Price'] = data1['Price'].apply(pricing)
```
