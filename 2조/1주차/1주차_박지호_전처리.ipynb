### 2. Installs: string -> int

```python
# string 2 int 함수
def size_2_int(size):
    if size.isdigit() == True:
        size = int(size)
    elif 'M' in size:
        size = size.replace('M', '')
        size = int(float(size) * 1000000)
    elif 'k' in size:
        size = size.replace('k','')
        size = int(float(size) * 1000)
    return size
```

```python
# Size: string -> int
data_info['Size'] = data_info['Size'].map(size_2_int)
```
