```python
import numpy as np
import pandas as pd
```

```python
appdata = pd.read_csv('../archive_googleplaystore/googleplaystore.csv')
review = pd.read_csv('../archive_googleplaystore/googleplaystore_user_reviews.csv')
```

## app data 전처리

```python
appdata = appdata.replace('Varies with device', np.nan)
appdata = appdata.drop('Type', axis=1)
appdata['Installs'] = np.array(x[:-1] for x in list(appdata['Installs']))
```

```python
def size_converter(x):
    if 'M' in x:
        return(float(x[:-1])*1000000)
    elif 'k' in x:
        return(float(x[:-1])*1000)
    else:
        pass

def price_converter(x):
    if x != '0':
        return (float(x[1:]))
    else:
        return(0)
def install_converter(x):
    return float(x.replace(',',''))
```

```python
appdata['Size']= appdata['Size'].apply(size_converter)
appdata['Price'] = appdata['Price'].apply(price_converter)
appdata['Installs'] = appdata['Installs'].apply(install_converter)
```

## review data 전처리

```python
#결측치 제거 (Review가 NaN이면 나머지도 결측값)
review = review.dropna(subset=['Translated_Review']).reset_index(drop=True)
review
```

```python
review2= review.groupby('App').mean()
review2
```

```python
review_apps = review2.index.tolist()
app_with_reviews=[]
for name in review_apps:
    if name in appdata['App'].unique().tolist():
        app_with_reviews.append(name)
len(app_with_reviews)
```

```python
for x in range(0,len(appdata)):
    drop_lst=[]
    if appdata.App[x] not in app_with_reviews:
        drop_list.append(x)
appdata.drop(index=x).reset_index()
```
