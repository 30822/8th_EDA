```python
'''
같은 맥락으로, ‘ratings_disabled’열이 True라면, ‘likes’, ‘dislikes’ 모두 0일 것임. 이것 또한 사실인지 확인해보기.
'''
df.loc[df['ratings_disabled'] == True,['likes', 'dislikes']].head(n=5)
```

```python
'''
열 ‘publishedAt’, 열 ‘trending_date’ 둘 다 object 형태임을 위에서 확인하였으니, 
이를 모두 datatime64[ns] 형태로 변환하기. 
print(df[['publishedAt','trending_date']].head(5)) 를 실행하였을 때 아래와 같이 출력되어야 함. 
또한 자료형을 출력하여, datatime64[ns]인지 다시 확인.
'''

df['publishedAt'] = pd.to_datetime(df['publishedAt'].dt.tz_localize(None),format='%Y-%m-%d %H:%M:%S') #pd.to_datetime(df['date'])
df['trending_date'] = pd.to_datetime(df['trending_date'].dt.tz_localize(None),format='%Y-%m-%d %H:%M:%S') 

print(df[['publishedAt','trending_date']].head(5))
print('-'*30)
print('\n')
print(df[['publishedAt','trending_date']].info())
```

```python
'''
열 ‘categoryId’를 int type이 아닌 category type로 변환하기.
'''

df['categoryId'] = df['categoryId'].astype('category')

#df.info()
```

```python
'''
3. cat, df의 categoryId와 관련한 문제 해결해보기!
1) df에서 각 category가 몇 개씩 있는지 출력하고, 
2) 전체 중에서의 각 category가 차지하는 비율 출력하기
'''

print(df['categoryId'].value_counts())

print('-'*30)

print(df['categoryId'].value_counts(normalize=True))
```

```python
'''
cat에서의 열 ‘id’와 df의 열 ‘categoryId’가 같은 내용임을 확인할 수 있으므로(데이터 소개 확인) 데이터를 합치기. 
최종 결과물은 df로 저장하며, cat에서의 ‘snippet.title’ 열이, df의 ‘category_name’열이 되도록 합치기. 
'''

df = pd.merge(df,cat,left_on = 'categoryId', right_on = 'id',how='left')
df
```
