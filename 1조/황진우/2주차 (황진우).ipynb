## 첫번째 분석

* 분석한 내용 간단히 설명 : 연령에 따른 tp05(Open to new experiences, complex.)를 분석

* 연령이 어릴수록 tp05값이 높을 것으로 판단
* 빈도 수가 아닌 비율로 파악
* 나이에 따른 tp05값의 차이는 없는 것으로 판단
* 나이와 상관없이 대체적으로 tp05는 우하향하는 모습을 보인다

```python
#결측치 여부 판단
data.head()
```

```python
data.head()
```

```python
pivot1 = pd.pivot_table(data = data, index = "tp05", columns = "age_group", aggfunc = "size", fill_value = 0)
pivot1
```

```python
for i in pivot1.columns:
    pivot1[i] = pivot1[i]/pivot1[i].sum()
```

```python
pivot1
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = range(8)
for c in pivot1.columns:
    y = pivot1[c].values
    plt.plot(x, y, label = c)
    
plt.legend()   
plt.show()
```

## 세번째 분석

```python
df.head()
```

```python
df.iloc[:,1:24:2]
```

```python
df['E'] = df.iloc[:,1:24:2].sum(axis=1)
```

```python
df.head()
```

```python
a = []
for i in range(1,4):
    a.append((df['urban'] == i).sum())
    print(i, a[i-1])
```

```python
df[['urban','E']].groupby('urban').sum()
```

```python
group = df[['urban','E']].groupby('urban').sum()
```

```python
for i in range(3):
    group.iat[i,0] = group.iat[i,0] / a[i]
```

```python
group['E']
```

```python
ls = ['rural','suburban','urban']
plt.bar(ls, group['E'])
plt.show()
```

```python
import seaborn as sns
from IPython.display import Image
```

```python
Answers = ['QaA', 'QbA', 'QcA', 'QdA', 'QeA',
             'QfA', 'QgA', 'QhA', 'QiA', 'QjA', 
             'QkA', 'QlA', 'QmA', 'QnA', 'QoA', 
             'QpA', 'QqA', 'QrA', 'QsA', 'QtA']
```

```python
dat = pd.read_csv("data.csv")
```

```python
correlations = dat[Answers].corr(method = 'spearman')
sns.heatmap(correlations, cmap="coolwarm", square=True, center=0)
```

```python
for col in dat[Answers]:
    print(sorted(dat[col].unique()))
```

```python
flipping_columns = ["QeA", "QfA", "QkA", "QqA", "QrA"]
for flip in flipping_columns: 
    dat[flip] = 6 - dat[flip]
```

```python
correlations = dat[Answers].corr(method = 'spearman')
sns.heatmap(correlations, cmap="coolwarm", square=True, center=0)
```

```python
flipping_secret_columns = ["QaA", "QdA", "QgA", "QiA", "QnA"]
for flip in flipping_secret_columns: 
    dat[flip] = 6 - dat[flip]
```

```python
correlations = dat[Answers].corr(method='spearman')
sns.heatmap(correlations, cmap="coolwarm", square=True, center=0)
```

```python
dat['Mach_score'] = dat[Answers].mean(axis = 1)
dat.head()
```

```python
dat[['Mach_score']].describe()
sns.displot(dat, x="Mach_score", bins=100)
```

```python
dat = dat[dat.urban != 0]
```
