##3. 혼인 상태 (△)

여성이 남성에 비해 혼인한 상태일 확률이 더 높을 것으로 예상되었다. 특히, 성격에서 비판적이고 다투기 좋아하는 성격에 강력하게 동의하는 비율이 남성이 여성에 비해 2배 정도 높았기 때문에 남성은 결혼 상태를 유지하기 어려울 것이라고 예상되었다.

```python
pivot12 = pd.pivot_table(data = data, index = "married", columns = "gender", aggfunc = "size", fill_value = 0)
pivot12['Female_percent'] = (pivot12['Female']/pivot12['Female'].sum()) * 100
pivot12['Male_percent'] = (pivot12['Male']/pivot12['Male'].sum()) * 100
pivot12 = pivot12.drop([0], axis=0)
pivot12
```

한번도 혼인을 하지 않았다고 응답한 비율이 남성이 여성에 비해 약 9%가 높은 것으로 나타났고, 혼인한 상태라고 응답한 비율은 여성이 2%, 이전에 혼인했던 경험이 있다고 응답한 비율은 여성이 6%비율로 높았다. 따라서, 크게 유의미한 결과라고 보기는 어렵지만 어느 정도는 남성이 여성에 비해 초혼 비율이 떨어진다는 것을 알 수 있다.

```python
pivot29 = pd.pivot_table(data = data, index = "urban", columns = "gender", aggfunc = "size", fill_value = 0)
pivot29['Female_percent'] = (pivot29['Female']/pivot29['Female'].sum()) * 100
pivot29['Male_percent'] = (pivot29['Male']/pivot29['Male'].sum()) * 100
pivot29 = pivot29.drop([0], axis=0)
pivot29
```

시골 또는 지방에 살았다고 답한 비율은 모두 19%로 여성과 남성의 비율이 일치하였고, 교외에 산 사람들은 각각 40%와 41%로 1% 차이밖에 나타나지 않았으며, 마지막으로 도시에 살았다고 답한 사람들은 여성이 39%, 남성이 38%로, 이도 마찬가지로 1% 비율 차이만 나타났다. 따라서, 성별에 따라 유년기의 거주 구역은 차이가 나타나지 않음을 알 수 있다.

###5.9 지난 해 국가 선거 투표 여부 (X)

투표 여부도 예상과 일치하게 성별 간 차이가 나타나지 않았다. 투표에 참여한 비율은 여성이 2% 높고 참여하지 않은 비율은 2% 낮았을 뿐, 큰 유의미한 차이는 존재하지 않는 것으로 분석된다.
