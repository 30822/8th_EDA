```python
import requests
from bs4 import BeautifulSoup 
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
```

```python
url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=101'
res = requests.get(url, headers = headers)
res.raise_for_status()
soup = BeautifulSoup(res.text, 'lxml') 
a1 = soup.findAll('a', attrs = {'class' : 'cluster_text_headline nclicks(cls_lif.clsart)'})
```

```python
for a in a1:
    title = a.get_text()
    link = a['href']
    print(f'기사 제목 : {title}') 
    print(f'(링크 : {link})')
```

```python
for i in range(5):
    url = f'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=101#&date=%2000:00:00&page={i}'
    res = requests.get(url, headers = headers)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml') 
    a2 = soup.findAll('a',attrs = {'class': 'cluster_text_headline nclicks(cls_lif.clsart)'})
    
    
    for a in a2:
        title = a.get_text().strip()
        link = a['href']
        print(f'기사 제목 : {title}') 
        print(f'(링크 : {link})')
```
